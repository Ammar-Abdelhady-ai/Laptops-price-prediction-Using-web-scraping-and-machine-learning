```python
laptops_pattern = r"\W\w+\W\w+\W\w+\W\w?\d+\w\d?\w?|\W\w+\W\w+\W\w+\W\w?\d+\w\d?\w?|\w+®?\W\w+™?\W\w\d\w\d+\w|\w+®?\W\w+™?\W\w\d|\w+\W\w+\W\d|\w+\W\d?\w+\W\w?\d|\w\d-?\w\d\w\d+\w?\d?|\w+\W\w+™?\W?\d+\W?\d+\w?|\w+®\W\w+®"

def fix_processor(row):
    is_exit = True
    try:
        data = re.findall(laptops_pattern, row)
        for item in data:
            pros = re.findall(r"i\d", item)
            if pros:
                return item
                is_exit = False

            elif "intel" in item.lower():
                return item
                is_exit = False
                
            elif "amd" in item.lower():
                return item
                is_exit = False
                
            elif "celeron" in item.lower():
                return item
                is_exit = False
                
            elif re.findall(r"\d\d\d\d\w", item):
                return item
                is_exit = False
                
        if is_exit:
            return np.nan
            
    except:
        return np.nan
        
    
```

```python
fill_processor = df[df["CPU Type"].isna()]["title"].apply(fix_processor).tolist()
fill_processor
```

```python
if len(fill_processor) == len(df[df["CPU Type"].isna()]["title"]):
    
    for processor, index in zip(fill_processor, df[df["CPU Type"].isna()]["title"].index):
        df["CPU Type"].iloc[index] = processor
```

```python
ssd_pattern = r"\d+.?.b.?.?.?.?.?.?.?.?.?.?.?.?.?.?.?.?.?ssd?|\d+.?.?.?.?.?solid|\d+.?.?.?.?.?.?hybrid.?drive"

def fix_ssd(row):
    row = str(row).lower()
    try:
        
        if "ssd"  in row or "solid" in row:
            return re.findall(ssd_pattern, row)[-1]
        
        elif "hybrid drive" in row or "ssd" in row:
            return re.findall(ssd_pattern, row)[-1]
        
        return 0
            
    except:
        return 0
```

```python
hdd_pattern = r"\d+.?.b.?.?.?.?.?.?.?.?.?.?.?.?hdd|\d+.?.?.?.?.?hard.?disk"

def fix_hdd(row):
    row = str(row).lower()
    try:
        
        if "hdd" in row or "hard disk" in row:
            return re.findall(hdd_pattern, row)[-1].split("+")[-1]
        
        return 0
            
    except:
        return 0
```

```python
def fix_hdd_GB(row):
    row = str(row).lower()
    try:
        
        if "tb" in row:
            item = re.findall(r"\d\W?tb", row)[-1]
            return int(re.findall(r"\d", item)[-1]) *1000
        
        else:
            return re.findall(r"\d+", row)[-1]
        
    except:
        return row
```

```python
def fix_ssd_GB(row):
    
    row = str(row).lower()
    try:
        if "tb" in row:
            item = re.findall(r"\d.?tb", row)[-1]
            return int(re.findall(r"\d", item)[-1])*1000
            
        elif "." in row:
            return re.findall(r"\d+", row.split(",")[-1])[0]          
            
        else:
            return re.findall(r"\d+", row)[-1]
    
    except:
        return row
```
