```python
def fix_hard_ssd(row):
    row = str(row).lower()
    
    if "ssd" in row or "solid" in row:
        data = re.findall(r"\d+", row)
        if "ram" in row:
            return data[-2]
            
        elif len(data) > 1:
            return data[-1]
        
        else:
            return data[0]
    else:
        return 0
```

```python
df["ssd"] = df["con_hard"].apply(fix_hard_ssd)
df["ssd"].tolist()
```

```python
def fix_hard_hdd(row):
    row = str(row).lower()
    
    try:
        if "ssd" in row or "solid"  in row:
            return 0
        
        else:
            data = re.findall(r"\d+", row)
            return data[0]
        
    except:
        return 0
```

```python
df["hdd"] = df["con_hard"].apply(fix_hard_hdd).tolist()
df["hdd"]
```
