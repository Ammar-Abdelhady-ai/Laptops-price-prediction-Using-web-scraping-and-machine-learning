#### we use the regular expression to find the feature pattern

```python
laptops_pattern = r"\W\w+\W\w+\W\w+\W\w?\d+\w\d?\w?|\W\w+\W\w+\W\w+\W\w?\d+\w\d?\w?|\w+®?\W\w+™?\W\w\d\w\d+\w|\w+®?\W\w+™?\W\w\d|\w+\W\w+\W\d|\w+\W\d?\w+\W\w?\d|\w\d-?\w\d\w\d+\w?\d?|\w+\W\w+™?\W?\d+\W?\d+\w?|\w+®\W\w+®"

def fix_processor(row):
    is_exit = True
    try:
        data = re.findall(laptops_pattern, row)
        for item in data:
            pros = re.findall(r"i\d", item)
            if pros:
                return item
                is_exit = False

            elif "intel" in item.lower():
                return item
                is_exit = False
                
            elif "amd" in item.lower():
                return item
                is_exit = False
                
            elif "celeron" in item.lower():
                return item
                is_exit = False
                
            elif re.findall(r"\d\d\d\d\w", item):
                return item
                is_exit = False
                
        if is_exit:
            return np.nan
            
    except:
        return np.nan
        
    
```

```python
df["processor"] = df["info"].apply(fix_processor)
df["processor"]
```

### find hard and split it into two type (SSD, HDD)

```python
hard_pattern = r"\d+.?gb.?ssd?|\d+t.+ssd?|\d+gb.+ssd|\d+gb.storage|\d+.+solid|\d+.+mm.|\d+.?tb...|\d+.+gb.pc.+ssd|\d+.?gb.?ssd|\d+.?nvm|\d+g.?ssd|\d\d\dgb|\d+.?nvm|\d+g.?ssd|\d\d\dgb|\d+.?M.?2"

def fix_hard_ssd(row):
    is_exit = True
    hard = " + ".join(re.findall(hard_pattern, str(row).lower()))
    
    try:
        data = re.findall(r"\d+.+", hard)[0].strip().replace("and", ",").replace("+", ",").replace("–", ",").replace("-", ",").split(",")
        
        for i in data:
            if "ssd" in i:
                is_exit = False
                return i
            
            if "nvm" in i or "m.2" in i:
                is_exit = False
                return i
            
        if is_exit:
            return 0
    except:
        return 0
```

```python
df["ssd"] = df["info"].apply(fix_hard_ssd)
df["ssd"].value_counts()
```

```python
hard_pattern = r"\d+.?gb.?ssd?|\d+t.+ssd?|\d+gb.+ssd|\d+gb.storage|\d+.+solid|\d+.+mm.|\d+.?tb...|\d+.+gb.pc.+ssd|\d+.?gb.?ssd"

def fix_hard_hdd(row):
    is_exit = True
    hard = " + ".join(re.findall(hard_pattern, str(row).lower()))
    
    try:
        data = re.findall(r"\d+.+", hard)[0].strip().replace("+", ",").replace("–", ",").replace("and", ",").replace("-", ",").split(",")
        
        for i in data:
            
            if "hd" in i:
                is_exit = False
                return i
            
            
            elif "tb" in i and "ssd" not in i:
                is_exit = False
                return i
            
            elif re.findall(r"\d\d\dgb", i) and "ssd" not in i:
                is_exit = False
                return i
            
        if is_exit:
            return 0
    except:
        return 0
```

```python
df["hdd"] = df["info"].apply(fix_hard_hdd)
df["hdd"].value_counts()
```

```python
def fix_hard_GB(row):
    row = str(row).lower()
    
    if "tb" in row:
        return int(re.findall(r"\d+", row)[0]) * 1000
    return int(re.findall(r"\d+", row)[0])
```

```python
df["hdd"] = df["hdd"].apply(fix_hard_GB)
df["hdd"].unique()
```
