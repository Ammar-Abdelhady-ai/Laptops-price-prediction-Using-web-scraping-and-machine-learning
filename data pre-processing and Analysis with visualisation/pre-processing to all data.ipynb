#### operating to Arabic words

```python
translator = Translator()
arab_gpu = []


for i in df[(df["brand"] == "لاب") | (df["brand"] == "ديل")]["Graphics Card"]:
    translation = translator.translate(f"{i}", dest="en")
    print(translation.text)
    arab_gpu.append(translation.text)
```

```python
arab_brand = []

for i in df[(df["brand"] == "لاب") | (df["brand"] == "ديل")]["brand"]:
    translation = translator.translate(f"{i}", dest="en")
    print(translation.text)
    arab_brand.append(translation.text)
```

```python
# replace arabic words to english

if len(arab_brand) == len(df[(df["brand"] == "لاب") | (df["brand"] == "ديل")]) == len(arab_gpu):
    
    for arab_gpu, arab_brand, index in zip(arab_gpu, arab_brand, df[(df["brand"] == "لاب") | (df["brand"] == "ديل")].index):
        df["brand"].iloc[index] = arab_brand
        df["Graphics Card"].iloc[index] = arab_gpu
```

```python
# replace capital letters to small letters

def fix_object_columns(row):
    return str(row).lower()
```

```python
df["brand"] = df["brand"].apply(fix_object_columns)
df["Processor"] = df["Processor"].apply(fix_object_columns)
df["Graphics Card"] = df["Graphics Card"].apply(fix_object_columns)
df["store"] = df["store"].apply(fix_object_columns)
```

```python
# delet symbol from cpu text

def fix_cpu(row):
    return re.sub(r'[^\w\s]|\d*\W*MHz', '', row).split("processor")[0].split("8core")[0].split("turbo")[0]
```

```python
df["Processor"] = df["Processor"].apply(fix_cpu)
df["Processor"].tolist()
```

```python
# delet symbol from gpu text

def fix_gpu(row):
    return re.sub(r'[^\w\s]|\d*\W*MHz', '', row)
```

```python
df["Graphics Card"] =  df["Graphics Card"].apply(fix_gpu)
df["Graphics Card"].tolist()
```
