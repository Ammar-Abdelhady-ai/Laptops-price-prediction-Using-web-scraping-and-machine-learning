```python
def concat_columns(df, cols_to_concat, new_col_name, sep):
    df[new_col_name] = df[cols_to_concat[0]]
    
    for col in cols_to_concat[1:]:
        df[new_col_name] = df[new_col_name].astype(str) + sep + df[col].astype(str)
    
    global l 
    l = []
    for i in df[new_col_name]:
        l.append(i.replace("nan", "").strip())
        
    df[new_col_name] = l
    
    return df[new_col_name]
```

```python
hard_pattern = r"\d+.?gb.?ssd?|\d+t.+ssd?|\d+gb.+ssd|\d+gb.storage|\d+.+solid|\d+.+mm.|\d+.?tb...|\d+.+gb.pc.+ssd"

def fix_hard_ssd(row):
    hard = " ; ".join(re.findall(hard_pattern, str(row).lower()))
    data = ''
    
    try:
        data = re.findall(r"\d+.+", hard)[0].split(",")[-1].split("ram")[-1].split("|")[-1].split("-")[-1].split(" ; ")

        if len(data) == 2:
            if "tb" in data[0]:
                return int(re.findall(r"\d+", data[0].strip())[-1]) * 1000
            return int(re.findall(r"\d+", data[0].strip())[-1])
            

        elif "hd" in data[0].strip():
            return 0
            
        else:
            if "tb" in data[0]:
                return int(re.findall(r"\d+", data[0].strip())[-1]) * 1000
            return int(re.findall(r"\d+", data[0].strip())[-1])
            
    except:
        return 0
```
